```python
# Import packages 
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
# Import the url

# Create a variable to store the url
url = 'https://www.worldometers.info/world-population/population-by-country/'

# Create a variable to store the information 
page = requests.get(url)

# Make contact with the website
page
```

Connection established and ok to proceed

```python
# Get the information from the website.
if page.status_code == 200:
    html_doc = page.text # Use the attribute text

# At this point, we have got our html code. 

# Look at the html code.
# Create a variable to store the HTML info.
# Parse in the html_doc.
# Create an object 'soup'. Use BeautifulSoup constructor to pass in html_doc.
soup = BeautifulSoup(html_doc)

# Print the output in a readable format.
print(soup.prettify())
```

```python
# Try to find the table by ID first
table = soup.find('table', attrs={'id': 'example2'})

# If not found, try by common class name
if table is None:
    table = soup.find('table', class_='datatable w-full border border-zinc-200 datatable-table')

# If still not found, just grab the first <table> on the page
if table is None:
    table = soup.find('table')

# Print or raise an error if nothing is found
if table:
    print(table.prettify())
else:
    print("Could not find a table on this page.")
```

```python
# All of the rows of the table.
rows = table.find_all('tr')

# View the rows.
rows
```

```python
# Store the extracted data.
# Iterate through each of the rows and each of the values.
# Create an empty list to store our processed data.
output = []

# Specify the column names.
column_names = ['ID', 'Country (or dependency)', 'Population (2020)',
                'Yearly Change', 'Net Change', 'Density (P/Km2)',
                'Land Area (Km2)', 'Migrants (net)', 'Fert. Rate',
                'Med. Age', 'Urbn Pop', 'World Share']

# Create a for loop statement.
for country in rows:
    country_data = country.find_all('td')
    if country_data: # Short-hand for instance country_data has stuff in it.
        # Extract the text within each element. Convert tags into text.
        country_text = [td.text for td in country_data]
        # Store data in a zip format for easy access. Zip takes two lists of equal length, and zip them together into a list of tuples.
        output.append(dict(zip(column_names, country_text))) # cast the zip into a dictionary.
        
# View the result.
output
```

```python
# Create a DataFrame directly from the output.
data = pd.DataFrame(output)

# View the DataFrame.
data.head()
```

```python
# Save the DataFrame as a CSV file without index.
data.to_csv('countries.csv', index=False)
```

```python
# Create a JSON file.
import json

# Create a JSON file.
output_json = json.dumps(output)

# View the output.
output_json
```

```python
# Save the JSON file to .json.
# open an object in memory and write to it - come up with f.
with open('countries.json', 'w') as f:
    json.dump(output, f)
```
